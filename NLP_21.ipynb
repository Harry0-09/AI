```python
import torch
import torch.nn as nn
import torch.functional as F
```

```python
def masked_mean_pool(last_hidden, att_mask):
  input_mask_exp = att_mask.unsqueeze(-1).expand(last_hidden.size()).float()

  sum_embed = torch.sum(last_hidden * input_mask_exp, dim=1)

  sum_mask = torch.clamp(input_mask_exp.sum(dim=1), min=1e-9)

  avg = sum_embed / sum_mask

  return F.normalize(avg)
```
