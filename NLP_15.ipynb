```python
import torch
import torch.nn as nn
import math
```

```python
class ScaledDotProductAttention(nn.Module):
  def __init__(self, model_dim, d_k, d_q, d_v):
    super().__init__()
    self.W_Q = nn.Linear(model_dim, d_q)
    self.W_K = nn.Linear(model_dim, d_k)
    self.W_V = nn.Linear(model_dim, d_v)
    self.d_k = d_k

  def forward(self, x):
    Q = self.W_Q(x)
    K = self.W_K(x)
    V = self.W_V(x)

    score = (torch.matmul(Q, K.transpose(-2,-1))) / math.sqrt(self.d_k)

    seq_len = x.size(1)

    mask = torch.triu(torch.ones(seq_len, seq_len), diagonal=1).bool()

    scores = score.masked_fill(mask, -1e9)

    alfa = torch.softmax(scores, dim=-1)

    return torch.matmul(alfa, V)

```
