```python
import torch
import torch.nn as nn
```

```python
class SimpleRNN(nn.Module):
  def __init__(self, vocab_size, emb_dim, hidden_dim):
    super().__init__()
    self.embed = nn.Embedding(vocab_size, emb_dim)
    self.U = nn.Linear(hidden_dim, hidden_dim, bias=False)
    self.W = nn.Linear(emb_dim, hidden_dim)
    self.V = nn.Linear(hidden_dim, vocab_size)

  def forward(self, x):
    batch_size, seq_len = x.shape

    h = torch.zeros(batch_size, self.U.in_features)

    outputs = []

    for t in range(seq_len):
      e = self.embed(x[:, t])
      h = torch.relu(self.U(h)+self.W(e))
      y = self.V(h)

      outputs.append(y)

    return (torch.stack(outputs, dim=1), y)

rnn = SimpleRNN(256, 32, 64)
```
