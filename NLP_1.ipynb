```python
import re
import unicodedata
```

```python
class TextCleaner():

  def clean_html(self, text):
    new_text = re.sub("<[^>]+>", "", text)
    return new_text

  def normalize_unicode(self, text):
    final = ""
    for i in text:
      num = ord(i)
      if num <= 32 or num == 127:
        continue
      final += i

    return final

  def preserve_features(self, text):
    hash = re.findall(r"#\w+", text)
    url = re.findall(r"https?:\S+", text)

    return hash, url


with open("text.txt", "r") as f:
  tekst = f.read()

cleaner = TextCleaner()

#print(cleaner.clean_html(tekst))
#print(cleaner.normalize_unicode(tekst))
out1, out2 = cleaner.preserve_features(tekst)

print(f"hash: {out1}")
print(f"url: {out2}")
```
